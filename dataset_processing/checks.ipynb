```python
import os
import pandas as pd
from pathlib import Path
from collections import defaultdict
```

```python
DATASET_PATH = Path(os.path.abspath('')).parent / 'dataset'
RAW_PATH = DATASET_PATH / 'raw'
SPLITTED_PATH = DATASET_PATH / 'splitted'
DT_COL = 'event_time'
CHUNKSIZE = 1000000
```

```python
dfs = []
for f in os.listdir(RAW_PATH):
    if f.startswith('sample'):
        continue
    else:
        df = pd.read_csv(RAW_PATH / f, parse_dates=[DT_COL])
        df['date'] = df['event_time'].dt.date
        dfs.append(df)
df = pd.concat(dfs)
result = df.groupby(['date', 'event_type', 'category_code']).agg(Count=("price", "count")).reset_index()
```
